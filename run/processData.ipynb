```python
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.faker import Faker
from pyecharts.commons.utils import JsCode

data = [
        {"value": 0.33,"contract_address":"0xssdsdsdsdsd","tx_hash":"zxxxsdsd"},
        {"value": 0.27, "contract_address": "0xssdsdsdsdsd", "tx_hash": "zxxxsdsd"},
        {"value": 0.88, "contract_address": "0xssdsdsdsdsd", "tx_hash": "zxxxsdsd"},
        {"value": 0.67, "contract_address": "0xssdsdsdsdsd", "tx_hash": "zxxxsdsd"},
        {"value": 0.37, "contract_address": "0xssdsdsdsdsd", "tx_hash": "zxxxsdsd"},
    ]


bar = (
        Bar()
        .add_js_funcs("console.log('js function')")
        .add_xaxis(["12:23","12:24","12:25","12:26","12:27"])
        .add_yaxis(series_name="执行时间", y_axis = data)
        .set_global_opts(title_opts=opts.TitleOpts(title="合约执行时间"),
                         toolbox_opts=opts.ToolboxOpts(),
                         visualmap_opts=opts.VisualMapOpts(),
                         datazoom_opts=[
                             opts.DataZoomOpts(range_start=0, range_end=100),
                             opts.DataZoomOpts(type_="inside", range_start=0, range_end=100),
                         ],
                         axispointer_opts=opts.AxisPointerOpts(is_show=True),
                         xaxis_opts=opts.AxisOpts(name="开始执行时刻",name_rotate=-20),
                         yaxis_opts=opts.AxisOpts(name="合约执行时间"),
                         legend_opts=opts.LegendOpts(pos_left="center"),
                         tooltip_opts=opts.TooltipOpts(
                             formatter=JsCode(
                                 """
                                 function (params) {
                                     console.log(params);
                                     return '合约地址:'+ params.data.contract_address + '</br>' +
                                             '交易哈希:'+ params.data.tx_hash;
                                 }
                                 """
                             )
                         ),
             )

    )

bar.render("bar_markpoint_type.html")
```

```python
filepath="/Users/bethestar/Downloads/myFiscoBcos/fisco/nodes/127.0.0.1/node0/log_record/"
filename="net_p2p_transmission_latency.csv"

# 读取csv文件
df = pd.read_csv(filepath+filename)

# 取出send_id列的唯一值
send_id = df['send_id'].unique()[0]

# 打印send_id
print(send_id)

# 取出receive_id、duration两列
df = df[['receive_id', 'duration']]

# 去掉duration列末尾的's'
df['duration'] = df['duration'].str.rstrip('s')

# 将duration列转换为浮点数
df['duration'] = df['duration'].astype(float)

# 计算超时重传带来的误差
df['duration'] = df['duration'] - 3 * (df['duration'] // 3)

# 按receive_id分组，并计算duration的平均值
receive_id_avg = df.groupby('receive_id')['duration'].mean()

# 打印结果
print(receive_id_avg)
```
